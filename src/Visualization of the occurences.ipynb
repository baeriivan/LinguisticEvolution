```python
import seaborn
from __future__ import division
import matplotlib.pyplot as plt
%matplotlib inline
from dictFcts import *
from helpers import *
import pickle
from collections import Counter
```

```python
all_dict = load_dictionaries('data/dico/GDL/', 1804,1997)
```

```python
def get_occs(word,dictionaries):
    list_occ = []
    for dictionary in dictionaries:
        list_occ.append(dictionary[word])
    return list_occ
```

```python
years = range(1804,1998)
nb_articles_year = []
nb_words_year = []
for year in years:
    fileName = 'data/raw/GDL_pkl/' + str(year) + '.pkl'
    data = loadYear(fileName)
    nb_articles_year.append(len(data))
    nb = 0
    for article in data:
        nb += len(article)
    nb_words_year.append(nb)
    if year in [1850, 1900, 1950]:
        print("Data until " + str(year) + " retrieved")
print("All data are loaded")
```

```python
def plotOccurences(word, param):
    list_occs = get_occs(word, all_dict)
    if param == 'article':
        y = [a/b for a, b in zip(list_occs, nb_articles_year)]
    elif param == 'word':
        y = [a/b for a, b in zip(list_occs, nb_words_year)]
    else:
        y = list_occs
    fig, ax = plt.subplots()
    x = range(0,194)
    ax.set_xticks([0, 46, 96, 146, 193])
    ax.set_xticklabels(["1804", "1850", "1900", "1950", "1997"])
    ax.plot(x,y)
```

```python
def compareWordOcc(word):
    plotOccurences(word, '')
    plotOccurences(word, 'article')
    plotOccurences(word, 'word')
```
